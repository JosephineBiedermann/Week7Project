```python
# building a function to put the song ids in track_ids_latin into chunks
# the audio feature function can only process 50 tracks at the time

def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
# getting the audio features from the latin song ids
# iterating over the latin list with the chunk function

audiofeats_latin1 = []
for chunk in list(chunks(tracks_ids_flat, 50)):
    audiofeats_latin1.append(sp.audio_features(chunk))
    
audiofeats_latin1
```

```python
# turning the audio feature list into a df

import pandas as pd

pd.DataFrame(audiofeats_latin1)

# looks like there are dict in the cells
```
